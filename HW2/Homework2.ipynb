```python
import pandas as pd
```

We start by reading data from a csv-file and add a new column with the prices per square meters for each apartment.

```python
df = pd.read_csv("Booli_sold.csv")
```

```python
df["ppsqm"] = df["soldPrice"] / df["livingArea"]
df
```

Here we create a new table sorted in prices per square meters descending. From which we can create another table with the top 5 apartments.

```python
df_sort_ppsqm = df.sort_values("ppsqm", ascending=False)
df_top5_ppsqm = df_sort_ppsqm.head(5)
df_top5_ppsqm
```

To calculate the average price per square meters for all the 158 apartments we sum over column "ppsqm" and divide by the number of rows. We get that the average price per square meter is around 58 000.

```python
avg_ppsqm = df["ppsqm"].sum() / len(df["ppsqm"])
avg_ppsqm
```

If we again take the table sorted in price per square meters but look at the top 20 apartments and then sort by "soldDate" we can see that the oldest apartment is sold 2015-07-02. We can also see the index 89 furthest to the left of this apartment representing the place in the original dataframe *df* which was sorted by most recently sold. 

Since the index numbers started at 0 for the most recently sold apartment we have that the apartment sold 2015-07-02 was the 90th most recently sold and since this was the apartment sold longest time ago in the top 20 price per square meters we can draw the conclusion that none of the 20 most expensive apartments per square meter was sold before 2015-07-02.

In other words, none of the 68 apartments sold longest time ago was one of the 20 most expensive per square meters. This shows a trend of increasing prices which confirms what we have all been hearing in the last decade about increasing costs. This can of course depend on other factors as well, size of the apartments sold etc... But still interesting to me.

## The Swedish Election of 2018
Our next assignment is about the Swedish election 2018 and so we read the data from a csv-file but notice that the data is separated by semicolon and values having a comma instead of decimal point..

```python
df_election = pd.read_csv("2018_R_per_kommun.csv", sep=";", decimal = ",")
df_election
```

We want to see the total number of legitimate votes in Stockholm so we can simplify the dataframe with only two columns and sum up legitimate votes over the municipalities in Stockholm. This lets us easily see that the number of legitimate votes are 1 426 237.

```python
df_election_cleaned = df_election[["LÄNSNAMN", "RÖSTER GILTIGA"]]
df_election_cleaned.groupby("LÄNSNAMN").sum()
```

The municipality where the social democratic party had its highest voting percentage was Munkfors with 52.18%, which can be seen below.

```python
df_election.sort_values("S", ascending=False)
```

To see which 3 municipalities had the highest participation we can sort the dataframe along the column "VALDELTAGANDE". To make it more readable we focus on just the two necessary columns, "KOMMUNNAMN" and "VALDELTAGANDE". Below we see that top 3 participation was in Lomma, Habo and Vellinge.

```python
df_election_participation = df_election.sort_values("VALDELTAGANDE", ascending=False)
df_election_participation[["KOMMUNNAMN", "VALDELTAGANDE"]].head(3)
```

## Predicting Strokes


```python
import matplotlib.pyplot as plt
```

To draw some conclusion about stroke we can read from a csv file with data on different parameters which might play a role in stroke risk.

```python
df_strokes = pd.read_csv("stroke-data.csv")
df_strokes
```

From this table we can for example look at the relation of age and stroke prevalence by calculating the mean for strokes per age group and plot the relation.

```python
df_strokes_age = df_strokes[["age", "stroke"]]
df_strokes_age_mean = df_strokes_age.groupby("age").mean()
plt.plot(df_strokes_age_mean)
```

In the plot we can see there is a clear increase in stroke prevalence with increasing age which is to be expected.
