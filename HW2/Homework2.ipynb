```python
import pandas as pd
```

We start by reading data from a csv-file and add a new column with the prices per square meters for each apartment.

```python
df = pd.read_csv("Booli_sold.csv")
```

```python
df["ppsqm"] = df["soldPrice"] / df["livingArea"]
df
```

Here we create a new sorted table with the 5 apartments having the highest price per square meters.

```python
df_top5_ppsqm = df.sort_values("ppsqm", ascending=False).head(5)
df_top5_ppsqm
```

Now we calculate the average price per square meters for all the 158 apartments.

```python
avg_ppsqm = df["ppsqm"].sum() / len(df["ppsqm"])
avg_ppsqm
```

If we look again at the price per square meters but expand the table to the top 20 apartments and then sort by oldest we can see that no apartment sold before 2015-07-02 makes the list. 

```python
df_top20_ppsqm = df.sort_values("ppsqm", ascending=False).head(20)
df_top20_ppsqm.sort_values("soldDate")
```

From the dataframe *df* which is sorted by most recently sold we can see that the apartment sold 2015-07-02 is on place 90. Therefore none of the 68 oldest apartments makes the top 20. This shows a clear trend of increasing prices.

## The Swedish Election of 2018

Our next assignment is about the Swedish election 2018 and so we read the data from a csv-fil but notice the data is separated by semicolons..

```python
df_election = pd.read_csv("2018_R_per_kommun.csv", sep=";")
df_election
```

We want to see the total number of legitimate votes in Stockholm so we can simplify the dataframe with only two columns. This lets us easily see that the number of legitimate votes are 1 426 237.

```python
df_election_cleaned = df_election[["LÄNSNAMN", "RÖSTER GILTIGA"]]
df_election_cleaned.groupby("LÄNSNAMN").sum()
```

The municipality where the social democratic party had its highest voting percentage was Munkfors with 52.18%, which can be seen below.

```python
df_election.sort_values("S", ascending=False)
```
