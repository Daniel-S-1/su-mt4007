We start by reading data from a csv-file and add a new column with the prices per square meters for each apartment.

```python
df = pd.read_csv("Booli_sold.csv")
```

```python
df["ppsqm"] = df["soldPrice"] / df["livingArea"]
```

Here we create a new sorted table with the 5 apartments having the highest price per square meters.

```python
df_sort_by_ppsqm = df.sort_values("ppsqm", ascending=False).head(5)
df_sort_by_ppsqm
```

Now we calculate the average price per square meters for all the apartments.

```python
avg_ppsqm = df["ppsqm"].sum() / len(df["ppsqm"])
avg_ppsqm
```
